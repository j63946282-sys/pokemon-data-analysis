```python
df['Type 2'] = df['Type 2'].fillna('None')
```

```python
df = df.drop_duplicates()
```

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df['Is_Fast'] = df['Speed'] > 100
df.head()
```

```python
df.sort_values('Total', ascending=False).head(10)
```
